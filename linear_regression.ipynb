now using the sklearn choose the model selection based on reducing the errors 

```python
x = df.drop(['sales','total_spent'], axis=1)
x
```

```python
y = df['sales']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
                                    x, y, test_size=0.33, random_state=101)
```

```python
X_train.head
```

```python
X_test.head
```

```python
from sklearn.linear_model import LinearRegression
#help(LinearRegression)
```

```python
len(df)#check for the data frame rows and after split by below command
```

```python
len(X_train)
```

```python
len(y_train)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
test_pred = model.predict(X_test)
test_pred
```
